# Gradient descendent project

```
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt
```

First let's graph our cost function

```
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})   #we create the graph defining the 3d plane

def f(x,y):
    return np.sin(x) + 2*np.cos(y)

res = 100
x = np.linspace(-4,4,num = res)
y = np.linspace(-4,4,num = res)

x, y = np.meshgrid(x,y)

z = f(x,y)

surf = ax.plot_surface(x,y,z, cmap = cm.cool)

fig.colorbar(surf)
```

Second, we need the level map

```
level_map = np.linspace(np.min(z),np.max(z),num = res)
plt.contourf(x,y,z,levels = level_map,cmap = cm.cool)
plt.colorbar()
plt.title('Gradient descendent')

p = np.random.rand(2) * 8 - 4  #to create a random vector from -4 to 4 of 2 dimensions that why we multiply for 8 and rest for 4
plt.plot(p[0],p[1], 'o', c='k')

h = 0.01            #increment random
lr = 0.01

def derivate(cp,p):          #we create a function to calculate the derivates
    return (f(cp[0],cp[1]) - f(p[0],p[1])) / h

def gradient(p):
    grad = np.zeros(2)         #we create a vector with ceros initially
    for idx, value in enumerate(p):
        cp = np.copy(p)               #to copy the initial point p
        cp[idx] = cp[idx] + h         #we sum the increment to the vector

        dp = derivate(cp,p)             #we calculate the partial derivates
        grad[idx]=dp                  #we save the gradient calculated
    
    return grad

for i in range(1000):
    p = p - lr*gradient(p)
    if(i%10 == 0):
        plt.plot(p[0],p[1], 'o', c='r')
plt.plot(p[0],p[1], 'o', c='w')         #last point
```
